Прилад містить 7 елементів, з яких 3 зношені. При включені приладу випадковим чином включаються три елементи.
Знайти ймовірність того, що при цьому виявляться включеними два зношених елементи.

Вирішення:

1. Прилад випадковим чином може включити 3 елементів наступною кількістю способів:

$${P(C)} = {C}^{3}_{7} = \frac{7!}{4! * 3!} = \frac{4! * 5 * 6 * 7}{4! * 2 * 3} = \frac{5 * 7}{1} = 35$$

2. Дві браковані деталі можуть включитися наступною кількістю способів:

$${P(A)} = {C}^{2}_{3} = \frac{3!}{1! * 2!} = \frac{2! * 3}{1! * 2!} = \frac{3}{1} = 3$$

3. Одна ціла деталь може включитися наступною кількістю способів:

$${P(B)} = {C}^{1}_{4} = 4$$

4. Наступною кількістю способів можна включити дві браковані і одну цілу деталь:

$${P(A)} * {P(B)} = 3 * 4 = 12$$

5. За класичним визначенням, дві браковані деталі і одну цілу включити можна з наступною вірогідністю:

$${P(D)} = \frac{P(A) * P(B)}{P(C)} = \frac{12}{35}$$
